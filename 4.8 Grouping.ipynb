## Aggregating data with transform

```python
# Grouping, transform, and new column
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## Deriving Columns with loc()

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] > 10) & (df_ords_prods_merge['max_order'] <=40), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

## Aggregating data with transform for spenders

```python
# Grouping, transform, and new column
df_ords_prods_merge['average_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.head(10)
```

```python
df_ords_prods_merge.tail(10)
```

#### It looks like average prices are mostly less than 10.

### Deriving Columns with loc()

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
df_ords_prods_merge['spender_flag'].value_counts(dropna=False)
```

#### We see that the number of High spenders is only near 2%. 

### Aggregating data with transform for frequency

```python
# Grouping, transform, and new column
df_ords_prods_merge['median_freq'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.head(5)
```

### Create 'freq_Flag' 

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_freq'] > 20, 'freq_Flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_freq'] > 10) & (df_ords_prods_merge['median_freq'] <= 20), 'freq_Flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_freq'] <= 10, 'freq_Flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge.head(5)
```

```python
df_ords_prods_merge['freq_Flag'].value_counts(dropna=False)
```
